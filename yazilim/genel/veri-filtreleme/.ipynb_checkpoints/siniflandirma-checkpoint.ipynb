```python
import pandas as pd
```

```python
file_path = "genel-ayrilmis.xlsx"
df = pd.read_excel(file_path)
filtered_df = df.copy()
```

```python
filtered_df = filtered_df[filtered_df["egitim"].isin(["Üniversite", "Yüksek Lisans - Doktora"])]

# cinsiyet için kategoriler
cinsiyet_map = {
    "Erkek": 1, "Kadın": 2
}
filtered_df["cinsiyet"] = filtered_df["cinsiyet"].map(cinsiyet_map).fillna(filtered_df["cinsiyet"])

# eğitim için kategoriler
egitim_map = {
    "Üniversite": 1, "Yüksek Lisans - Doktora": 2
}
filtered_df["egitim"] = filtered_df["egitim"].map(egitim_map).fillna(filtered_df["egitim"])

# sporsure için kategoriler
bedensporsure_map = {
    "0": 1, "1-30": 2, "31-60": 3,
    "61-90": 4, "91-120": 4, "121+": 4
}
filtered_df["bedensporsure"] = filtered_df["bedensporsure"].map(bedensporsure_map).fillna(filtered_df["bedensporsure"])

# sporgun için kategoriler
bedensporgun_map = {
    "0": 1, "1": 2, "2": 3,
    "3-4": 4, "5-6": 4, "7": 4
}
filtered_df["bedensporgun"] = filtered_df["bedensporgun"].map(bedensporgun_map).fillna(filtered_df["bedensporgun"])

# yoruculuk için kategoriler
bedenyoruculuk_map = {
    "0": 1, "1-2": 2, "3-4": 3, "5-6": 4, "7-8": 4, "9-10": 4
}
filtered_df["bedenyoruculuk"] = filtered_df["bedenyoruculuk"].map(bedenyoruculuk_map).fillna(filtered_df["bedenyoruculuk"])

# sportakim için kategoriler
bedensportakim_map = {
    "0": 1, "1-2": 2, "3-4": 2, "5-6": 2, "7-8": 2, "9-10": 2
}
filtered_df["bedensportakim"] = filtered_df["bedensportakim"].map(bedensportakim_map).fillna(filtered_df["bedensportakim"])

# sporsure için kategoriler
zihinsporsure_map = {
    "0": 1, "1-30": 2, "31-60": 3,
    "61-90": 4, "91-120": 4, "121+": 4
}
filtered_df["zihinsporsure"] = filtered_df["zihinsporsure"].map(zihinsporsure_map).fillna(filtered_df["zihinsporsure"])

# sporgun için kategoriler
zihinsporgun_map = {
    "0": 1, "1": 2, "2": 3,
    "3-4": 4, "5-6": 4, "7": 4
}
filtered_df["zihinsporgun"] = filtered_df["zihinsporgun"].map(zihinsporgun_map).fillna(filtered_df["zihinsporgun"])

# yoruculuk için kategoriler
zihinyoruculuk_map = {
    "0": 1, "1-2": 2, "3-4": 3, "5-6": 4, "7-8": 4, "9-10": 4
}
filtered_df["zihinyoruculuk"] = filtered_df["zihinyoruculuk"].map(zihinyoruculuk_map).fillna(filtered_df["zihinyoruculuk"])

# sportakim için kategoriler
zihinsportakim_map = {
    "0": 1, "1-2": 2, "3-4": 2, "5-6": 2, "7-8": 2, "9-10": 2
}
filtered_df["zihinsportakim"] = filtered_df["zihinsportakim"].map(zihinsportakim_map).fillna(filtered_df["zihinsportakim"])

# not için kategoriler
not_map = {
    "0-25": 1, "26-40": 1, "41-55": 1, "56-70": 1,
    "71-85": 2, "86-100": 3
}
filtered_df["not"] = filtered_df["not"].map(not_map).fillna(filtered_df["not"])

# derszamani için kategoriler
derszamani_map = {
    "0": 1, "1-3": 1, "4-8": 2, "9-14": 3, "15-20": 4, "21+": 4
}
filtered_df["derszamani"] = filtered_df["derszamani"].map(derszamani_map).fillna(filtered_df["derszamani"])

# derskatilim için kategoriler
derskatilim_map = {
    "0": 1, "1-2": 2, "3-4": 3, "5-6": 4, "7-8": 4, "9+": 4
}
filtered_df["derskatilim"] = filtered_df["derskatilim"].map(derskatilim_map).fillna(filtered_df["derskatilim"])

# kisiselplan için kategoriler
kisiselplan_map = {
    "0": 1, "1-2": 1, "3-4": 2, "5-6": 3, "7-8": 4, "9-10": 5
}
filtered_df["kisiselplan"] = filtered_df["kisiselplan"].map(kisiselplan_map).fillna(filtered_df["kisiselplan"])

# derszaman için kategoriler
derszaman_map = {
    "0": 1, "1-2": 1, "3-4": 1, "5-6": 2, "7-8": 2, "9-10": 3
}
filtered_df["derszaman"] = filtered_df["derszaman"].map(derszaman_map).fillna(filtered_df["derszaman"])

# okularkadas için kategoriler
okularkadas_map = {
    "0": 1, "1-2": 1, "3-4": 1, "5-6": 2, "7-8": 3, "9-10": 4
}
filtered_df["okularkadas"] = filtered_df["okularkadas"].map(okularkadas_map).fillna(filtered_df["okularkadas"])

# grupcalisma için kategoriler
grupcalisma_map = {
    "0": 1, "1-2": 1, "3-4": 2, "5-6": 2, "7-8": 3, "9-10": 3
}
filtered_df["grupcalisma"] = filtered_df["grupcalisma"].map(grupcalisma_map).fillna(filtered_df["grupcalisma"])

# dersstresi için kategoriler
dersstresi_map = {
    "0": 1, "1-2": 1, "3-4": 2, "5-6": 3, "7-8": 4, "9-10": 4
}
filtered_df["dersstresi"] = filtered_df["dersstresi"].map(dersstresi_map).fillna(filtered_df["dersstresi"])

# dersyorgunluk için kategoriler
dersyorgunluk_map = {
    "0": 1, "1-2": 1, "3-4": 1, "5-6": 2, "7-8": 3, "9-10": 3
}
filtered_df["dersyorgunluk"] = filtered_df["dersyorgunluk"].map(dersyorgunluk_map).fillna(filtered_df["dersyorgunluk"])

# liderlik için kategoriler
liderlik_map = {
    "0": 1, "1-2": 1, "3-4": 2, "5-6": 3, "7-8": 4, "9-10": 5
}
filtered_df["liderlik"] = filtered_df["liderlik"].map(liderlik_map).fillna(filtered_df["liderlik"])

# ozguven için kategoriler
ozguven_map = {
    "0": 1, "1-2": 1, "3-4": 1, "5-6": 2, "7-8": 3, "9-10": 4
}
filtered_df["ozguven"] = filtered_df["ozguven"].map(ozguven_map).fillna(filtered_df["ozguven"])
```

```python
output_file = "siniflandirilmis-veriler.csv"
filtered_df.to_csv(output_file, index=False, sep=',')
```
