```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "siniflandirilmis-veriler-bedensel.csv"  # Filtrelenmiş dosyanızın adını yazın
df = pd.read_csv(file_path)
```

```python
correlation_with_not = df.corr()["not"][["yas", "cinsiyet", "egitim", "sporgun", "sporsure", "yoruculuk", "sportakim", "sporamac"]]
```

```python
correlation_df = correlation_with_not.to_frame()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_df, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
```

```python
plt.title("Bağımsız Değişkenlerin 'not' ile Korelasyonu")
plt.show()
```
