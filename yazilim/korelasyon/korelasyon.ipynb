```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "filtrelenmis-veriler.csv"
df = pd.read_csv(file_path)
```

```python
correlation_with_not = df.corr()["not"][["yas", "sporgun", "sporsure", "sportakim"]]
```

```python
correlation_df = correlation_with_not.to_frame()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_df, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
```

```python
plt.show()
```
