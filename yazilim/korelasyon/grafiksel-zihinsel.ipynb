```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "siniflandirilmis-veriler-zihinsel.csv"
df = pd.read_csv(file_path)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Bar grafikleri çiz
sns.barplot(x="yas", y="not", data=df, ax=axes[0, 0], palette="Blues")
axes[0, 0].set_title("Yaş - Not İlişkisi")

sns.barplot(x="cinsiyet", y="not", data=df, ax=axes[0, 1], palette="Reds")
axes[0, 1].set_title("Cinsiyet - Not İlişkisi")

sns.barplot(x="egitim", y="not", data=df, ax=axes[1, 0], palette="Greens")
axes[1, 0].set_title("Eğitim - Not İlişkisi")

sns.barplot(x="sporgun", y="not", data=df, ax=axes[1, 1], palette="Purples")
axes[1, 1].set_title("Spor Gün Sayısı - Not İlişkisi")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.barplot(x="sporsure", y="not", data=df, ax=axes[0, 0], palette="Oranges")
axes[0, 0].set_title("Spor Süresi - Not İlişkisi")

sns.barplot(x="yoruculuk", y="not", data=df, ax=axes[0, 1], palette="coolwarm")
axes[0, 1].set_title("Spor Yoruculuk - Not İlişkisi")

sns.barplot(x="sportakim", y="not", data=df, ax=axes[1, 0], palette="BuGn")
axes[1, 0].set_title("Takım Sporu - Not İlişkisi")

sns.barplot(x="sporamac", y="not", data=df, ax=axes[1, 1], palette="viridis")
axes[1, 1].set_title("Spor Amaç - Not İlişkisi")
```
